Use a for loop to create a list of such names.  You can use the `.count()` string method to figure out whether a name has enough e's (s.count(n) returns the number of times n appears in s). Copy your code from the last section to iterate through all the combinations of two names with two other names (again just using names with at least two e's), but this time, keep track of which combination has the most e's (uppercase or lowercase) between all four names. In case of a tie, you only need to print one of the best combinations.

````python
best_count = 0
all_four = []

for first in both_names:
  for second in both_names:
    if first != second:
      four = (first + second)
      all_four.append(four)

for combo in all_four:
    number = sum(name.lower().count('e') for name in combo)
    if number > best_count:
        best_count = number
        best_combo = (combo, number)   # store the best combination

print("Best names:", best_combo)






````
